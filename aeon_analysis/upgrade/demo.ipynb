```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
root = Path(r"C:\Users\neurogears\Desktop\JonNewman_2019-12-13T09_07_16")
linear_acc_path  = root / "linear-accel_2019-12-13T09_07_16.raw"
gravity_path = root / "gravity-vector_2019-12-13T09_07_16.raw"
euler_path = root / "euler_2019-12-13T09_07_16.raw"
quarterion_metadata_path = root / "quaterion-time_2019-12-13T09_07_16.csv"
```

```python
## Load Metadata including timestamps
colnames=['Timestamp', 'Counter']
metadata = pd.read_csv(quarterion_metadata_path, names=colnames, header=None)
metadata["Timestamp"] = pd.to_datetime(metadata["Timestamp"])


## Load Linear Acceleration Data
linear_acc = np.fromfile(linear_acc_path, dtype=np.dtype(np.float64))
linear_acc = np.reshape(linear_acc, (-1, 3))
plt.figure()
plt.plot(metadata["Timestamp"].values, linear_acc, label=["x", "y", "z"])
plt.xlabel("Timestamp")
plt.ylabel("Acceleration (m/s^2)")
plt.title("Linear Acceleration")
plt.legend()
plt.show()

## Load Gravity Vector Data
gravity_vec = np.fromfile(gravity_path, dtype=np.dtype(np.float64))
gravity_vec = np.reshape(gravity_vec, (-1, 3))
plt.figure()
plt.plot(metadata["Timestamp"].values, gravity_vec, label=["x", "y", "z"])
plt.xlabel("Timestamp")
plt.ylabel("Acceleration (m/s^2)")
plt.title("Gravity Vector")
plt.legend()
plt.show()

## Load Gravity Vector Data
euler_angle = np.fromfile(euler_path, dtype=np.dtype(np.float64))
euler_angle = np.reshape(euler_angle, (-1, 3))
plt.figure()
plt.plot(metadata["Timestamp"].values, euler_angle, label=["yaw", "pitch", "roll"])
plt.xlabel("Timestamp")
plt.ylabel("Angle (deg)")
plt.title("Orientation (Euler Angles)")
plt.legend()
plt.show()

print((len(metadata), len(linear_acc), len(gravity_vec), len(euler_angle)))
```
