```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import cv2
```

```python
def variance_of_laplacian(image):
	# compute the Laplacian of the image and then return the focus
	# measure, which is simply the variance of the Laplacian
    lap = cv2.Laplacian(image, cv2.CV_64F, ksize=3)
    return (lap, lap.var(), )
```

```python
im_path = r"C:\Users\neurogears\Desktop\JonsData\223132933-eb4493f4-3190-4e29-adfb-21b4840fc5e8.png"
img = cv2.imread(im_path, 0)
imgs = [img, img]
```

```python
for i in imgs:
    i = cv2.equalizeHist(i)
    lap, var = variance_of_laplacian(i)
    lap = cv2.convertScaleAbs(lap)

    plt.figure()
    plt.imshow(lap, cmap='gray')
    plt.title(f"Variance: {var:.5f}")
    plt.colorbar()
    plt.show()
```
